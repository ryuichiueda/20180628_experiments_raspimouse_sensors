```python
import pandas as pd       ###hist_600###
import matplotlib.pyplot as plt
import glob
import re

df = pd.DataFrame()

for f in glob.glob("../*mm.txt"): 
    data  = pd.read_csv(f, delimiter=" ", header=None, names = ("date","time","rf","rs","ls","lf"))

    rng = re.sub(r'\D', '', f)
    
    df[rng] = [data["lf"].mean(), data["lf"].std(),data["rf"].mean(), data["rf"].std()]
    del data
```

```python
result = df.transpose()
result.columns = ["left_mean","left_stddev","right_mean","right_stddev"]
result.to_csv("result.csv", sep=",")
result
```
